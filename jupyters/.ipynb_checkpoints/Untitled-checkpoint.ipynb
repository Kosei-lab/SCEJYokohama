```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.read_excel("averageForce_forGraph.xlsx")
```

```python
matplotlib.font_manager.weight_dict["roman"]
matplotlib.font_manager._rebuild()
```

```python
time = df["Time[s]"]
```

```python
df.head()
```

```python
drag = df["drag"].values
pressure = df["pressure"].values
gravity = df["gravity"].values
```

```python
drag
```

```python
def plotGraphs():
  
    c_linestyle="solid"
    c_color="black"
    c_markerfacecolor="None"
    c_markersize=10
    
#     plt.plot(time, f100, "o-", linestyle=c_linestyle, markerfacecolor=c_markerfacecolor, markersize=c_markersize,label="100 mm/s")
#     plt.plot(time, f300, "o-", linestyle=c_linestyle, markerfacecolor=c_markerfacecolor, markersize=c_markersize,label="300 mm/s")
#     plt.plot(time, f500, "o-", linestyle=c_linestyle, markerfacecolor=c_markerfacecolor, markersize=c_markersize,label="500 mm/s")
#     plt.plot(time, f700, "o-", linestyle=c_linestyle, markerfacecolor=c_markerfacecolor, markersize=c_markersize,label="700 mm/s")
    
    plt.rcParams['font.family'] = 'IPAexMincho'
    plt.plot(time, drag, "o", color="blue", linestyle=c_linestyle, markerfacecolor=c_markerfacecolor, markersize=c_markersize,label="流体抗力")
    plt.plot(time, pressure, "x", color="red", linestyle=c_linestyle, markerfacecolor=c_markerfacecolor, markersize=c_markersize,label="圧力勾配による力")
    plt.plot(time, gravity, "^", color="black", linestyle=c_linestyle, markerfacecolor=c_markerfacecolor, markersize=c_markersize,label="重力")
    
    plt.legend(fontsize=14)
    
    plt.savefig("force.png", dpi=450)
    plt.show()
```

```python

fig = plt.figure(figsize=(8,6), dpi=450)
ax = fig.add_subplot(1,1,1)


ax.set_xlabel("Time [s]",fontsize=18)
ax.set_ylabel("Force [N]",fontsize=18)
ax.set_xlim([0, 0.020])
ax.set_ylim([-1.0e-7, 3.0e-7])

plt.xticks(np.arange(0, 0.021, 0.004))

ax.grid()


plt.rcParams['font.size'] = 12
#plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["xtick.direction"] = "in"
plt.rcParams["ytick.direction"] = "in"
plt.rcParams["axes.linewidth"] = 1.0

plt.tick_params(labelsize=14)

plotGraphs()
```
