```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.read_excel("flux.xlsx")
```

```python
matplotlib.font_manager.weight_dict["roman"]
matplotlib.font_manager._rebuild()
```

```python
time = df["Times [s]"]
```

```python
df.head()
```

```python
f300 = df[300].values
```

```python
f500 = df[500].values
```

```python
f700 = df[700].values
```

```python
f300
```

```python
f500
```

```python
def plotGraphs():
    plt.plot(time, f300, "o-",linestyle="solid", color="black",markerfacecolor="None",label="300 mm/s")
    plt.plot(time, f500, "^-",linestyle="dashed", color="black",markerfacecolor="None",label="500 mm/s")
    plt.plot(time, f700, "D-",linestyle="dashdot", color="black",markerfacecolor="None",label="700 mm/s")
    
    plt.legend(fontsize=14)
    
    plt.savefig("flux.png", dpi=500)
    plt.show()
    
```

```python

fig = plt.figure(figsize=(8,6), dpi=500)
ax = fig.add_subplot(1,1,1)


ax.set_xlabel("Time [s]",fontsize=14)
ax.set_ylabel(r"$\mathrm{Flux [1 / mm^2 \cdot s]}$",fontsize=14)
ax.set_xlim([0, 0.25])
ax.set_ylim([-1000, 25000])

plt.xticks(np.arange(0, 0.25, 0.04))

ax.grid()


plt.rcParams['font.size'] = 14
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["xtick.direction"] = "in"
plt.rcParams["ytick.direction"] = "in"
plt.rcParams["axes.linewidth"] = 1.0

plt.tick_params(labelsize=14)

plotGraphs()
```
