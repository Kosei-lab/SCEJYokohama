```python
def plotGraphs():
  
    c_linestyle="solid"
    c_color="black"
    c_markerfacecolor="None"
    c_markersize=10
    
    #plt.plot(time, f100, "o-", linestyle=c_linestyle, markerfacecolor=c_markerfacecolor, markersize=c_markersize,label="100 mm/s")
    #plt.plot(time, f300, "o-", linestyle=c_linestyle, markerfacecolor=c_markerfacecolor, markersize=c_markersize,label="300 mm/s")
    #plt.plot(time, f500, "o-", linestyle=c_linestyle, markerfacecolor=c_markerfacecolor, markersize=c_markersize,label="500 mm/s")
    #plt.plot(time, f700, "o-", linestyle=c_linestyle, markerfacecolor=c_markerfacecolor, markersize=c_markersize,label="700 mm/s")
    
    #plt.plot(time, f100, "o", linestyle=c_linestyle, markerfacecolor=c_markerfacecolor, markersize=c_markersize,label="100 mm/s")
    plt.plot(time, f300, "x", linestyle=c_linestyle, markerfacecolor=c_markerfacecolor, markersize=c_markersize,label="300 mm/s")
    plt.plot(time, f500, "^", linestyle=c_linestyle, markerfacecolor=c_markerfacecolor, markersize=c_markersize,label="500 mm/s")
    plt.plot(time, f700, "D", linestyle=c_linestyle, markerfacecolor=c_markerfacecolor, markersize=c_markersize,label="700 mm/s")
    
    plt.legend(fontsize=14)
    
    plt.savefig("flux.png", dpi=450)
    plt.show()
    
```

```python

fig = plt.figure(figsize=(8,6), dpi=450)
ax = fig.add_subplot(1,1,1)


ax.set_xlabel("Time [s]",fontsize=18)
ax.set_ylabel(r"$\mathrm{Flux [1\:/\:mm^2 \cdot s]}$",fontsize=18)
ax.set_xlim([0, 0.25])
ax.set_ylim([-1000, 25000])

plt.xticks(np.arange(0, 0.25, 0.04))

ax.grid()


plt.rcParams['font.size'] = 12
#plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["xtick.direction"] = "in"
plt.rcParams["ytick.direction"] = "in"
plt.rcParams["axes.linewidth"] = 1.0

plt.tick_params(labelsize=14)

plotGraphs()
```
